```
def f(t):
  return np.exp(-t)*np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1) #0~5까지 이거인듯
t2 = np.arange(0.0, 5.0, 0.02)
t1_ = torch.from_numpy(t1)
t2_ = torch.from_numpy(t2)

plt.figure()
plt.subplot(211) #subplot 공부하기
plt.plot(t1_, f(t1_), 'bo', t2_, f(t2_), 'k')

plt.figure()
plt.subplot(212)
plt.plot(t2_, np.cos(2*np.pi*t2), 'r--')

```

# Gradient 
*f* = 2*x^2 + 3*x +1  <br>
df/dx = 4*x+3 @5  <br>
f'(5) = 23

```
x = torch.tensor(5.0, requires_grad=True) #default는 false, true로 바꿔야 gradient 계산됨
f = 2*x**2 + 3*x + 1
f.backward()
x.grad #tensor(23.)
x.type() #torch.FloatTensor #무슨 과정인지 이해 안됨ㅎ
```

*f* = 2*x^2 + 3*y^2 +1  <br>
df/dx = 4*x @7  => 28<br>

df/dy = 6*y @8  => 48<br>



```
x = torch.tensor(7.0, requires_grad=True)
y = torch.tensor(8.0, requires_grad=True)

f = 2*x**2 + 3*y**2 + 1
f.backward()
print(x.grad)
print(y.grad)
```
