```
# import libraries
import torch
import numpy as np
import matplotlib.pyplot as plt
import torch.nn as nn
from sklearn import datasets
```

# generating training dataset


```
n_pts = 100

centers = [[1.0, 1.0], [-1.0, -1.0]]
#데이터개수,feature 개수,센터,표준편차
X, y = datasets.make_blobs(n_samples=n_pts, n_features=2, centers=centers, cluster_std=1.0)

x_data = torch.Tensor(X)
y_data = torch.Tensor(y.reshape(n_pts, 1))
#print(X.shape) (100,2), feature 2개!!
#print(y_data)
print(y.shape) #(100,)
```

# define scatter plot function

```
def scatterplot():
  plt.scatter(X[y==0, 0], X[y==0, 1], label='Negative(0) data') #데이터 2개, y=0인 첫번째 feature 두번째 feature data
  plt.scatter(X[y==1, 0], X[y==1, 1], label='Positive(1) data')
  #plt.scatter(X[y==2, 0], X[y==2, 1], label='Positive(2) data') #센터가 3개일때
  plt.legend(loc='best')

scatterplot() #왜 점이 많이 찍힘??
```

# Define model class
z = w1*x1 + w2*x2 + b  -> <br>
y = a = sigma(z) -> <br>
L(y_hat = a, y)


```
class LogisticRegression(nn.Module):
  def __init__(self, input_size, output_size):
    super().__init__()
    self.linear = nn.Linear(input_size, output_size) #2개,1개

  def forward(self, x): 
    pred = torch.sigmoid(self.linear(x))
    return pred   #probability (not direct value)

  def predict(self, x):
    pred = self.forward(x)
    if pred >= 0.5:
      return 1
    else:
      return 0


# instantiate model class
torch.manual_seed(1)
model = LogisticRegression(2, 1)
print(list(model.parameters()))
```

# function to get model parameters (w1, w2, b)

```
[w, b] = model.parameters()
#print(w) 
w1, w2 = w.view(2)

def get_params():
  return (w1.item(), w2.item(), b[0].item())
```

# show the initial fit plot

```
def fitplot():
  w1, w2, b = get_params()
  x1 = np.array([-3, 3])
  # 0 = w1x1 + w2x2 + b
  x2 = (w1*x1 + b)/(-w2)
  plt.plot(x1, x2, 'r')
  scatterplot()
  plt.show()


fitplot() #초기모델이라 피팅시켜야함...!
```

# training the model 

```
# instantitate optimizer 
criterion = nn.BCELoss()  # = nn.CrossEntropyLoss() # for LR with more than 2 classes
optimizer = torch.optim.SGD(model.parameters(), lr=0.02)

# training the model 
epochs = 1000
losses = []

for i in range(epochs):
  y_pred = model.forward(x_data)

  # calculrate loss 
  loss = criterion(y_pred, y_data)
  print("epoch: ", i, "loss: ", loss.item())
  losses.append(loss.item())

  optimizer.zero_grad() # clear gradients wrt parameters
  loss.backward()
  optimizer.step()
```

```
# log loss
plt.plot(range(epochs), losses)
plt.xlabel('loss')
plt.ylabel('epoch')
plt.grid()
```

```
fitplot() #학습 후 그래프
```

# model evalation with new datasets the model has never seen before

```
pnt1 = torch.Tensor([-0.1, -0.1]) #data point
pnt2 = torch.Tensor([2.0, 2.0])
plt.plot(pnt1.numpy()[0], pnt1.numpy()[1], 'rx', markersize=12)
plt.plot(pnt2.numpy()[0], pnt2.numpy()[1], 'kx', markersize=12)
fitplot()
print("red x (+, 1) prob= {}".format(model.forward(pnt1).item()))
print("red x's class = {}".format(model.predict(pnt1)))
print("blk x (+, 1) prob= {}".format(model.forward(pnt2).item()))
print("blk x's class = {}".format(model.predict(pnt2)))

```
